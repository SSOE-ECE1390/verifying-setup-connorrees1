```python
# Connor Rees - ECE 1390 - 8/26/2025
# Homework 0

# Imports
import cv2
import os
import mediapipe

import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load Image
apollo_image = cv2.imread('Data/Apollo_11_Launch.jpg')

# View Image
plt.imshow(apollo_image)
```

```python
print(apollo_image.shape)

# Crop Image
apollo_cropped = apollo_image[100:650, 500:750]
plt.imshow(apollo_cropped)
```

```python
# Brightness
matrix = np.ones(apollo_cropped.shape, dtype="uint8") * 72
apollo_darker = cv2.subtract(apollo_cropped, matrix)

plt.figure(figsize=[10, 5])
plt.subplot(131); plt.imshow(apollo_darker);  plt.title("Darker");
plt.subplot(132); plt.imshow(apollo_cropped); plt.title("Original");
```

```python
# Rotate
apollo_flipped = cv2.flip(apollo_cropped, -1)
plt.imshow(apollo_flipped)

# Save Image
cv2.imwrite("Modified_Apollo_Image.png", apollo_flipped)

# Removing clutter in github repo
os.remove("Modified_Apollo_Image.png")
```
